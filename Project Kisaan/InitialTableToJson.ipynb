```python
import json
import sys
```

```python
x_step = 500
x_max = 2500
y_step = 300
y_max = 1200
```

```python
obj = {u"gas": '',u"humidity": '',u"temperature": '',u"x": 'NaN',u"y": 'NaN',u"z": "NaN",u"time": ''}
```

```python
obj
```

```python
x_n = int(x_max / x_step)
y_n = int(y_max / y_step)
```

```python
data = []
for i in range(0,x_n):
    for j in range(0,y_n):
        data.append(obj)
```

```python
print(json.dumps(data, indent=4))
```

```python
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```
